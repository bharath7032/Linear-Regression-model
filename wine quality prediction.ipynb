```python
x=df.drop('quality',axis=1)
y=df['quality']
```

```python
#training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

```python
y_test
```

```python
#standardizing the training and testing data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#model fitting
from sklearn.linear_model import LinearRegression
rig=LinearRegression()
rig.fit(x_train,y_train)
```

```python
#predicting
y_pred=rig.predict(x_test)
```

```python
#importing and calculating metrics of linear regression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean Suared Error :",mse)
print("Mean Absolute Error :",mae)
print("R2 score :",r2)
```

```python
#scatter plot of actual vs predicted
plt.scatter(y_test,y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Linear Regression - Actaul vs Predicted")
plt.show()
```

```python
residuals=y_test-y_pred
residuals
```

```python
df_results=pd.DataFrame({'Acutal':y_test,'Predicted':y_pred})
print(df_results)
```

```python
mape=np.mean(np.abs((y_test-y_pred) / y_test))*100
accuracy=100-mape
print(f"Model Accuracy :{accuracy:.2f}%")
```

```python
#now predicting from the dataset by taking a row by index
sample=df.drop('quality',axis=1).iloc[2]
sample_scale=scaler.transform([sample])
prediction=rig.predict(sample_scale)
print("selected row features :\n",sample)
print("predicted quality : ",round(prediction[0],4))
```
